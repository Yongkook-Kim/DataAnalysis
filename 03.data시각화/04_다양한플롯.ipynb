# Matplotlib의 다양한 플롯

## 1. Bar Chart
- x데이터가 카테고리 값인 경우 사용

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)     # 레이블이 긴 경우 등등..
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

### 타이타닉 선실별 승객수

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
s = titanic['class'].value_counts()
s
```

```python
df = titanic[['class', 'pclass']].groupby('class').count()
df
```

```python
y = s.values
x = np.arange(len(y))
xlabel = s.index.to_list()

plt.title('타이타닉 선실별 승객 수')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.xlabel('선실')
plt.ylabel('승객수')
plt.show()
```

```python
y = df.pclass.values
x = np.arange(len(y))
xlabel = df.index.to_list()

plt.title('타이타닉 선실별 승객 수')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.xlabel('선실')
plt.ylabel('승객수', rotation=0, labelpad=15)
plt.show()
```

## 2. 파이 차트
- 카테고리 별 값의 상대적인 비교를 해야 할 때

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)            # 돼지가 약간 튀어나오도록 설정
plt.title("Pie Chart", fontsize=15)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)      # 90도에서 반시계 방향으로
plt.axis('equal')
plt.show()
```

### Anscombe's Quartet
- 선형회귀선: y = 3.0 + 0.5 * x

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.groupby('dataset').describe()
```

- Dataset I 에 대해서 산점도와 회귀선 그리기

```python
ds1 = ans[ans.dataset == 'I']
ds1
```

```python
plt.scatter(ds1.x, ds1.y)
plt.grid()
plt.title('Dataset I')
plt.show()
```

```python
plt.scatter(ds1.x, ds1.y)
plt.plot([2,16],[4,11], c='r', lw=2)
plt.grid()
plt.title('Dataset I')
plt.show()
```

- 4개의 데이터셋 모두에 대해서 그래프 그리기

```python
fig, axes = plt.subplots(2,2, figsize=(8,6))

for i, roman in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[i//2, i%2]
    ax.scatter(ds.x, ds.y)
    ax.grid()

plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(8,6),
                         sharex=True, sharey=True)

for i, roman in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[i//2, i%2]
    ax.scatter(ds.x, ds.y)
    ax.plot([2,20],[4,13], 'r-', lw=2)
    ax.set_title(f'Dataset {roman}')
    ax.grid()

fig.suptitle('Anscombe\'s Quartat', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

## 6. Box Plot

```python
iris = sns.load_dataset('iris')
iris.species.values
```

- Setosa 품종의 각 feature별 박스 플롯

```python
setosa = iris[iris.species == 'setosa']
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
             setosa.petal_length, setosa.petal_width],
            labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.title('Setosa 품종의 각 feature별 박스 플롯')
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(12,8),
                         sharex=True, sharey=True)

setosa = iris[iris.species == 'setosa']
plt.subplot(131)
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
             setosa.petal_length, setosa.petal_width],
            labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.title('Setosa 품종의 각 feature별 박스 플롯')
virginica = iris[iris.species == 'virginica']
plt.subplot(132)
plt.boxplot([virginica.sepal_length, virginica.sepal_width,
             virginica.petal_length, virginica.petal_width],
            labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.title('virginica 품종의 각 feature별 박스 플롯')
versicolor = iris[iris.species == 'versicolor']
plt.subplot(133)
plt.boxplot([versicolor.sepal_length, versicolor.sepal_width,
             versicolor.petal_length, versicolor.petal_width],
            labels=['sepal_length','sepal_width','petal_length','petal_width'])
plt.title('versicolor 품종의 각 feature별 박스 플롯')
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(12,8), sharey=True)

for i, species in enumerate(iris.species.unique()):
    sub = iris[iris.species == species]
    ax = axes[i]
    ax.boxplot([sub.sepal_length, sub.sepal_width, sub.petal_length, sub.petal_width],
            labels=['sepal_length','sepal_width','petal_length','petal_width'])
    ax.set_title(species)

fig.suptitle('Iris 모든 품종의 feature별 박스플롯', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
